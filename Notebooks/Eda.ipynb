```python
#Importinng Modulues
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#Loading the dataset
```

```python

df = pd.read_csv("Cleaned_Dataset.csv")
df.shape
df.info()
df.describe()

df['Revenue'] = df['Quantity'] * df['UnitPrice']
os.makedirs("results/Metrics", exist_ok=True)
```

```python
#Univariety Analysis
```

```python
print("Univariate Analysis...\n")
for col in ['Quantity', 'UnitPrice', 'Revenue']:
    plt.figure()
    sns.histplot(df[col], bins=50, kde=True)
    plt.title(f"{col} Distribution")
    plt.savefig(f"results/Metrics/{col}_distribution.png")
    plt.show()
```

```python
#Bivariety Analysis
```

```python
print("Bivariate Analysis...\n")

plt.figure()
sns.scatterplot(x='Quantity', y='Revenue', data=df)
plt.title("Quantity vs Revenue")
plt.savefig("results/Metrics/quantity_vs_revenue.png")
plt.show()

plt.figure()
sns.scatterplot(x='UnitPrice', y='Revenue', data=df)
plt.title("UnitPrice vs Revenue")
plt.savefig("results/Metrics/unitprice_vs_revenue.png")
plt.show()
```

```python
#Multivariety Analysis
```

```python
print("Multivariate Analysis...\n")

sns.pairplot(df[['Quantity', 'UnitPrice', 'Revenue']])
plt.savefig("results/Metrics/pairplot.png")
plt.show()
```

```python
# Top 10 Countries
```

```python
plt.figure()
df['Country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Countries")
plt.savefig("results/Metrics/top_10_countries.png")
plt.show()
```

```python
#Heatmap
```

```python
plt.figure()
sns.heatmap(df[['Quantity','UnitPrice','Revenue']].corr(), annot=True)
plt.title("Feature Correlation Heatmap")
plt.savefig("results/Metrics/heatmap.png")
plt.show()
```

```python
#Pattern Discovery
```
